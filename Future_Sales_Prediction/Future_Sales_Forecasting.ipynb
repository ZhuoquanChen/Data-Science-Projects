## - Facebook Prophet Model

```
# import prophet

from fbprophet import Prophet
```

```
# instantiate prophet model

# NOTE: rename two columns in data frame which are 'Date' column as 'ds'(x axis) and 'Sales' column as 'y'(y axis)

# 'store': specific store id
# 'sales_df': prediction result
# 'period': time period (days, weeks, months....)

def sales_predict(store, sales_df, period):
  
  sales_df = sales_df[sales_df['Store'] == store]
  sales_df = sales_df[['Date', 'Sales']].rename(columns={'Date':'ds', 'Sales':'y'})

  # instantiate prophet model
  model = Prophet()
  model.fit(sales_df)
  future = model.make_future_dataframe(period) # set up a period, like days, months....
  forecast = model.predict(future)

  return forecast
```

```
sales_predict(10, combined_df, 30)
```

```
# Visualizing the prediction

def sales_predict_plot(store, sales_df, period):
  
  sales_df = sales_df[sales_df['Store'] == store]
  sales_df = sales_df[['Date', 'Sales']].rename(columns={'Date':'ds', 'Sales':'y'})

  # instantiate prophet model
  model = Prophet()
  model.fit(sales_df)
  future = model.make_future_dataframe(period) # set up a period, like days, months....
  forecast = model.predict(future)

  # plot
  plot_figure_1 = model.plot(forecast, xlabel='Date', ylabel='Sales')
  plot_figure_1 = plt.grid(axis='x')
  plt.title('Forecasting The Sales Of Number 10 Store', fontdict={'fontsize': 15})
  plot_figure_2 = model.plot_components(forecast); # plot the components of forecast
```

```
sales_predict_plot(10, combined_df, 30) # in future 30 days (1 month)
```

**Training a Model With Holidays**

- StateHoliday: indicates a state holiday. Normally all stores, with few exceptions, are closed on state holidays. Note that all schools are closed on public holidays and weekends. a = public holiday, b = Easter holiday, c = Christmas, 0 = None
- SchoolHoliday: indicates if the (Store, Date) was affected by the closure of public schools

```
# Get all the dates pertaining(fitting) to school holidays 

school_holidays = combined_df[combined_df['SchoolHoliday'] == 1].loc[:, 'Date']
school_holidays
```

```
# Get all the dates pertaining to state holidays 
# the data frame indicates that it has three types of state holidays (a, b, c)

state_holidays = combined_df[(combined_df['StateHoliday'] == 'a') | (combined_df['StateHoliday'] == 'b')\
                             | (combined_df['StateHoliday'] == 'c')].loc[:, 'Date']
state_holidays
```

- swicth the data type of school_holidays and state_holidays (dtype: object) to dtype:datetime
- pd.to_datetime(state_holidays)

```
# create a school_holidays data frame

school_holidays_df = pd.DataFrame({'ds':pd.to_datetime(school_holidays), 'holiday':'school_holidays'})
school_holidays_df
```

```
# create a state_holidays data frame

state_holidays_df = pd.DataFrame({'ds':pd.to_datetime(state_holidays), 'holiday':'state_holidays'})
state_holidays_df
```

```
# concatenate both school and state holidays 
# columns have to be named as 'ds' and 'holiday', if not, the prophet model will not work.

combined_holidays = pd.concat([school_holidays_df, state_holidays_df])
combined_holidays
```

```
def sales_predict(store, sales_df, holidays, period):
  
  sales_df = sales_df[sales_df['Store'] == store]
  sales_df = sales_df[['Date', 'Sales']].rename(columns={'Date':'ds', 'Sales':'y'})

  # instantiate prophet model
  model = Prophet(holidays=holidays)
  model.fit(sales_df)
  future = model.make_future_dataframe(period) # set up a period, like days, months....
  forecast = model.predict(future)

  # plot
  plot_figure_1 = model.plot(forecast, xlabel='Date', ylabel='Sales')
  plot_figure_1 = plt.grid(axis='x')
  plt.title('Forecasting The Sales Of Number 10 Store With Holidays', fontdict={'fontsize': 15})
  plot_figure_2 = model.plot_components(forecast); # plot the components of forecast
```

```
# Making predictions using holidays for a specific store

sales_predict(9, combined_df, combined_holidays, 60)
```
